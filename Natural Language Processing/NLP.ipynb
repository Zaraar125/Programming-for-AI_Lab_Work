```python
text = """The feature on the new Pixel 7 is not as good Apple Inc’s Face ID unlocking mechanism, 
as it can struggle in low light and is more vulnerable to being spoofed. In addition, Google has 
said it is not secure enough to enable signing into apps or making payments.

The return comes after Google became stricter about launching products with facial recognition, 
in part due to questions about its performance on darker skin. The company took time to review its
approach to training and testing facial recognition since the previous Pixel with the capability 
launched in 2019, one of the sources said.

Google declined to comment on several specific questions about its history with face unlock. It 
said generally, “Thanks to advanced machine learning models for face recognition, Pixel 7 and 
Pixel 7 Pro feature Face Unlock, but we’re doing it a little differently.”"""
```

```python
import nltk
nltk.download('punkt')
import matplotlib.pyplot as plt
```

```python
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import numpy as np
```

```python
# Tokenize the sentence using nltk
# YOUR CODE HERE
tokenized_sentence=word_tokenize(text)
print(tokenized_sentence)
```

```python
# Create a frequency distributin using NLTK
# YOUR CODE HERE
fdist = FreqDist(i for i in tokenized_sentence)
print(fdist.most_common(len(fdist)))
```

```python
# Display the frequency distribution using matplotlib. Add proper x and y labels.
#YPUR CODE HERE
word=[]
rep=[]
for i,j in fdist.most_common():
  word.append(i)
  rep.append(j)
plt.figure(figsize=(20,30))
plt.barh(word,rep)
plt.xlabel("frequency")
plt.ylabel("words")
plt.show()
```

# Stop words removal

```python
nltk.download("stopwords")
```

```python
def remove_stopwords(word_list):
        processed_word_list = []
        n=stopwords.words("english")
        for i in word_list:
            if i.lower() not in n:
              processed_word_list.append(i)


        return processed_word_list
    
removed_stopwords = remove_stopwords(tokenized_sentence)
removed_stopwords
```

```python
# Create a frequency distributin using NLTK
# YOUR CODE HERE
fdist=FreqDist(i for i in removed_stopwords)
print(fdist.most_common())
```

```python
# Display the frequency distribution using matplotlib. Add proper x and y labels.
# YOUR CODE HERE
fdist=FreqDist(i for i in removed_stopwords)
word=[]
rep=[]
for i,j in fdist.most_common():
  word.append(i)
  rep.append(j)
plt.figure(figsize=(20,25))
plt.barh(word,rep)
plt.xlabel("frequency")
plt.ylabel("words")
plt.show()
```

```python
# Now remove the punctuation and create a frequency plot
def remove_punctuation(a):
  new_list=[i for i in a if i.isalnum()]
  return new_list

new_list=remove_punctuation(removed_stopwords)
fdist = FreqDist(i for i in new_list)
word=[]
rep=[]
for i,j in fdist.most_common():
  word.append(i)
  rep.append(j)
plt.figure(figsize=(20,30))
plt.barh(word,rep)
plt.ylabel("words without punctuation")
plt.xlabel("frequency")
plt.show()
```
