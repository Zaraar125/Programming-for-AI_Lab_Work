## Task 01

Your task is to merge all these files into a single dataframe. Columns will be as follows

dialog_id

dialog_text

dialog_act

dialog_emotion

```python
# Expected output of Task 01
```

```python
# YOUR CODE HERE
import pandas as pd
a=open("/content/i212705 Muhammad Zaraar Malik - dialogues_act.txt","r")
b=open("/content/i212705 Muhammad Zaraar Malik - dialogues_emotion.txt","r")
c=open("/content/dialogues_text.txt","r")


sample=c.read()
sample=sample.split("__eou__")
sample=[m.strip("\n") for m in sample]
sample1=b.read()
sample1=sample1.split(" ")
sample1=[o.strip("\n") for o in sample1]

sample2=a.read()
sample2=sample2.split(" ")
sample2=[s.strip("\n") for s in sample2]


n=pd.DataFrame(sample1)
n=n.head(102980)
m=pd.DataFrame(sample)
m=m.head(102980)
o=pd.DataFrame(sample2)
o=o.head(102980)

dialog_id=range(0,102980)

data=pd.concat([m,o,n],axis=1)
data.columns=["dialog_text","dialog_act",	"dialog_emotion"]
data["dialog_id"]=dialog_id

data.set_index("dialog_id",inplace=True)
data.head(5)
```

## Task 02

Next task is to assign label values to dialog_act and dialog_emotion as given in readme.txt

```python
# Expected output of Task 02
```

```python
# YOUR CODE HERE

data["dialog_act"]=data["dialog_act"].replace("1","inform")
data["dialog_act"]=data["dialog_act"].replace("2","question")
data["dialog_act"]=data["dialog_act"].replace("3","directive")
data["dialog_act"]=data["dialog_act"].replace("4","commisive")

data["dialog_emotion"]=data["dialog_emotion"].replace("0","no emotion")
data["dialog_emotion"]=data["dialog_emotion"].replace("1","anger")
data["dialog_emotion"]=data["dialog_emotion"].replace("2","disgust")
data["dialog_emotion"]=data["dialog_emotion"].replace("3","fear")
data["dialog_emotion"]=data["dialog_emotion"].replace("4","happiness")
data["dialog_emotion"]=data["dialog_emotion"].replace("5","sadness")
data["dialog_emotion"]=data["dialog_emotion"].replace("6","surprise")
data.head(5)   
    
```
