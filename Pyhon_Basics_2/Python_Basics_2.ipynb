## Task 03 - Create a class Campus with name, location, studentCount and facultyCount as attributes. 
1. Create a function names "warning" to generate a warning if count of students are more than 5 times of the count of faculty

2. Create a function to display all the object/instance values

```python
# YOUR CODE HERE
class Campus:
   def __init__(self,name,location,studentCount,facultyCount):
     self.name=[name]
     self.location=[location]
     self.studentCount=studentCount
     self.facultyCount=facultyCount
    
   def warnings(self):
     a=self.facultyCount*5
     if self.studentCount>a:
       print("warning the student count has exceded from the faculty count")

   def Display(self):
     print("Student names : ",self.name)
     print("Student locations : ",self.location)
     print("student count : ",self.studentCount)
     print("faculty count : ",self.facultyCount)

p1=Campus("zaraar","rawalpindi",10,1)
p1.warnings()
p1.Display()
```

## Task 05 - Find the longest number from "Numbers.txt"

```python
# YOUR CODE HERE
f=open("/content/Numbers.txt","rt")
a=f.read()
b=a.split()
c=[]
for i in b:
  c.append(int (i))
longest_word=0
for i in c:
  if longest_word<i:
    longest_word=i
print(longest_word)
```
